I am going to look at verbal vs math scores.  My assumption is that people are terrible at math so they will score higher in the verbal section.  I know they already stated which wasd higher but I like the programming challenge.

```python

verbal = np.array(data.iloc[:,11])
math = np.array(data.iloc[:,12])
```

```python
larger =[]
for i in range (0,len(math)):
    larger.append(math[i]>verbal[i])
```

```python
larger
```

```python
equal = []
for i in range(0,len(math)):
    equal.append(math[i]==verbal[i])
```

```python
equal
```

So I now have two arrays one showing when math is larger and the other showing when they are equal.  In this test we will throw away any data that is equal.

```python
sum(equal)
```

```python
sum(larger)
```

```python
len(math)
```

```python
min(sum(larger), len(math)-sum(larger)-sum(equal))
```

Here I am getting the smaller of the pluses and minuses.

```python
n = len(math)-sum(equal)
z = (119+.5-n/2)/(np.sqrt(n)/2)
```

```python
z
```

```python
stats.norm.cdf(z)
```

Well I can reject my null hypothesis!  Just my assumption was wrong...  More people did better on math than verbal...  I guess this class might agree with that any way...

There is a test in the stats package, let's see if it agrees with my result.

```python
stats.wilcoxon(math,verbal,correction=True)
```
